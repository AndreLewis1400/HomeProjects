```python
# Load the necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# Create the data points
x = np.array([0, 5, 3, 4, 7, 8, 10])
y = np.array([5, 2, 5, 15, 27, 15, 31])
```

```python
# Define the plot
sns.regplot(x=x, y=y, ci=None)

# Save the image
# plt.savefig("scatterplot.png")

# Show the image
plt.show()
```

```python
# Load the rock data set
df = pd.read_csv("https://static-resources.zybooks.com/DataScience/rock.csv")
```

```python
# Set title
plt.title('Area and perimeter of rock', fontsize=20)

# define plot
sns.regplot(x=df["Area"], y=df["Perimeter"], ci=None, fit_reg=False)

# show the image
plt.show()
```
